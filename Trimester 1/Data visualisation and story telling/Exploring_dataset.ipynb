```python
import pandas as pd

df = pd.read_csv('India_VAHAN_Dataset.csv')

# Key Metric 1: Total registrations by state
state_reg = df.groupby('State')['Registration_Number'].count().reset_index()
state_reg.columns = ['State', 'Total_Registrations']

# Key Metric 2: YoY Growth Rate
yearly = df.groupby('Registration_Year').size().reset_index(name='Count')
yearly['YoY_Growth_%'] = yearly['Count'].pct_change() * 100

# Key Metric 3: EV Adoption Rate by year
ev_rate = df.groupby('Registration_Year').apply(
    lambda x: (x['Fuel_Type']=='EV').sum() / len(x) * 100,
    include_groups=False
).reset_index(name='EV_Adoption_%')

# Key Metric 4: BS6 Compliance Rate
bs6_rate = df.groupby('Registration_Year').apply(
    lambda x: (x['Emission_Norm']=='BS6').sum() / len(x) * 100,
    include_groups=False
).reset_index(name='BS6_Compliance_%')

print(state_reg)
print(yearly)
print(ev_rate)
print(bs6_rate)
```
